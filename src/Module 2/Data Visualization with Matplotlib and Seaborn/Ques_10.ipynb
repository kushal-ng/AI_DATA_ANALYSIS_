```python
# Advanced Seaborn Plot with DataFrame
# Description: Create a Seaborn plot using a DataFrame that visualizes two features in a single plot. Use tips
# dataset available in Seaborn's sample datasets.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a Seaborn plot to visualize total_bill vs tip
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='sex', palette='coolwarm')

# Add title and labels
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill in($)')
plt.ylabel('Tip in($)')

# Display the plot
plt.show()
```
