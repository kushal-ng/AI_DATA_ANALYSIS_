```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
import matplotlib.pyplot as plt

# Data for product sales
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for height vs weight
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar plot for product sales
axes[0].bar(products, sales, color='skyblue')
axes[0].set_title('Product Sales (in thousands)')
axes[0].set_xlabel('Products')
axes[0].set_ylabel('Sales')

# Scatter plot for height vs weight
axes[1].scatter(heights, weights, color='green')
axes[1].set_title('Height vs Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')

# Display the plots
plt.tight_layout()
plt.show()
```
