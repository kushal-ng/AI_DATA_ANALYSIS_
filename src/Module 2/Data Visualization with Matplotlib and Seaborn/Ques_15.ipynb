```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
correlation_matrix = iris.select_dtypes(include=['float64', 'int64']).corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar_kws={'shrink': 0.8})

plt.title('Correlation Matrix Heatmap visualization')
plt.show()
```
