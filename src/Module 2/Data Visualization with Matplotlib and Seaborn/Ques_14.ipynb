```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.

import matplotlib.pyplot as plt

# Data for the line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a customized line plot
plt.plot(x, y, color='purple', linestyle='--', marker='o', markersize=8, linewidth=2)

# Add title and labels
plt.title('Customized Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X Axis', fontsize=12)
plt.ylabel('Y Axis', fontsize=12)

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```
