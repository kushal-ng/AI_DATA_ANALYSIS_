```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a Seaborn scatter plot to visualize the relationship between total_bill and tip
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Add title and labels
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill in ($)')
plt.ylabel('Tip in ($)')

# Display the plot
plt.show()
```
