```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.








# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.









# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.


import pandas as pd

# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.

# Load Iris dataset (assuming CSV file, or you can load from sklearn datasets)
iris_df = pd.read_csv('iris.csv')

# Check for missing values in the entire dataset
missing_values_total = iris_df.isnull().sum().sum()
print(f"Total missing values in the Iris dataset: {missing_values_total}")

print("\nMissing values per column:")
print(iris_df.isnull().sum())


# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.

# Sample data simulating inconsistent 'Gender' entries
sample_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['Female', 'M', 'male', 'Male', 'F']
}
sample_df = pd.DataFrame(sample_data)

# Normalize gender entries to lowercase for checking inconsistencies
sample_df['Gender_lower'] = sample_df['Gender'].str.lower()

# Find unique normalized gender entries
unique_genders = sample_df['Gender_lower'].unique()
print("\nUnique gender entries (normalized):")
print(unique_genders)

# Group original entries by normalized gender to identify variations
gender_groups = sample_df.groupby('Gender_lower')['Gender'].unique()
print("\nOriginal gender entries grouped by normalized gender:")
print(gender_groups)


# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.

print("\nIris Dataset Descriptive Statistics:")
print(iris_df.describe(include='all'))


```
