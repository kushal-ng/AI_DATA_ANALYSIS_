```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Customer Dataset
# - Load a custom CSV file named customer_data.csv .
# - Find any missing values in specific columns like 'Email' and 'Phone'.






# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Emails in Customer Dataset
# - Identify duplicate emails which might indicate duplicate customer records.







# Part 3: Generate a Data Quality Report

# Task 3: Customer Dataset Report
# - Summarize the data quality with missing values, duplicates, and inconsistencies for customer_data.csv .




import pandas as pd

# Part 1: Load a Dataset & Check Missing Values

# Task 1: Customer Dataset
# - Load a custom CSV file named customer_data.csv.
# - Find any missing values in specific columns like 'Email' and 'Phone'.

customer_df = pd.read_csv('customer_data.csv')

# Check missing values in 'Email' and 'Phone' columns
missing_email = customer_df['Email'].isnull().sum()
missing_phone = customer_df['Phone'].isnull().sum()

print(f"Missing Email values: {missing_email}")
print(f"Missing Phone values: {missing_phone}")


# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Emails in Customer Dataset
# - Identify duplicate emails which might indicate duplicate customer records.

# Find duplicate emails (keep first occurrence)
duplicate_emails = customer_df[customer_df.duplicated(subset=['Email'], keep=False)]

print("\nDuplicate Emails:")
print(duplicate_emails[['Email']].drop_duplicates())


# Part 3: Generate a Data Quality Report

# Task 3: Customer Dataset Report
# - Summarize the data quality with missing values, duplicates, and inconsistencies for customer_data.csv.

total_rows = len(customer_df)
total_duplicates = customer_df.duplicated().sum()
missing_per_column = customer_df.isnull().sum()

print("\nCustomer Dataset Quality Report:")
print(f"Total records: {total_rows}")
print(f"Total duplicate rows: {total_duplicates}")
print("Missing values per column:")
print(missing_per_column)

# For inconsistencies, example: check for inconsistent cases in Email (e.g., uppercase vs lowercase)
email_unique_case = customer_df['Email'].str.lower().nunique()
email_unique_original = customer_df['Email'].nunique()
if email_unique_case != email_unique_original:
    print("\nInconsistency detected: Emails have case variations that could cause duplicates.")

# Optionally list some inconsistent emails
inconsistent_emails = customer_df['Email'][customer_df['Email'].str.lower().duplicated(keep=False)]
if not inconsistent_emails.empty:
    print("\nSample emails with case inconsistencies:")
    print(inconsistent_emails.unique())

```
