```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Titanic Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the 'Age' column.







# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Rows in Titanic Dataset
# - Identify any duplicate rows in the dataset.










# Part 3: Generate a Data Quality Report

# Task 3: Titanic Dataset Overview
# - Create a simple report of missing values, duplicates, and some basic statistics for the Titanic dataset.




import pandas as pd

# Part 1: Load a Dataset & Check Missing Values

# Task 1: Titanic Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the 'Age' column.

# Load Titanic dataset (you can replace this path with your own)
titanic_df = pd.read_csv('titanic.csv')

# Check missing values in 'Age' column
missing_age_count = titanic_df['Age'].isnull().sum()
print(f"Missing values in 'Age' column: {missing_age_count}")


# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Rows in Titanic Dataset
# - Identify any duplicate rows in the dataset.

duplicates = titanic_df[titanic_df.duplicated()]
print(f"Number of duplicate rows: {duplicates.shape[0]}")
print("Duplicate rows preview:")
print(duplicates.head())


# Part 3: Generate a Data Quality Report

# Task 3: Titanic Dataset Overview
# - Create a simple report of missing values, duplicates, and some basic statistics.

def generate_data_quality_report(df):
    report = {}

    # Missing values per column
    report['missing_values'] = df.isnull().sum()

    # Total duplicates
    report['duplicate_rows'] = df.duplicated().sum()

    # Basic statistics for numerical columns
    report['basic_stats'] = df.describe()

    return report


quality_report = generate_data_quality_report(titanic_df)

print("\nData Quality Report:")
print("- Missing Values per Column:")
print(quality_report['missing_values'])
print(f"\nTotal Duplicate Rows: {quality_report['duplicate_rows']}")
print("\nBasic Statistics for Numerical Columns:")
print(quality_report['basic_stats'])

```
